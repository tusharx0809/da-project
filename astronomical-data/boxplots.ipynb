```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

star_df = pd.read_csv('https://drive.google.com/uc?id=1BQVc6MHjQFtDC9iP1isT_K4ojVe_Oil-')

star_df.head()
folder_name = 'star_plots'
base_dir = f'D:\\Python\\da-projects\\astronomical-data\\{folder_name}\\'
```

```python
plt.figure(figsize=(20,8))
plt.suptitle('Visualzing the outliers in numeric features of star type',
             color = 'blue',
             weight = 'bold')
for i in range(4):
  plt.subplot(1,4,i+1)
  sns.boxplot(x = star_df['Star type'], y = star_df.iloc[:,i]) #Here : represents all data and the number after comma is index
  plt.title(star_df.columns[i], color='red')
  plt.ylabel('')

plt.savefig(base_dir+'boxplot_startype.png')
```

```python
#Below code takes the country name as input and displays the years and gdp of that countries, we can compare for any countries

import pandas as pd
import matplotlib.pyplot as plt
import math
import random

#importing the excel file
excel_data = pd.read_excel('D:\\Python\\da-projects\\world-economic-indicator-project\\dataset\\WorldBank.xlsx')

def get_gdp_years(country):
  df = excel_data.loc[excel_data['Country Name'] == country, ['Year', 'GDP (USD)']]
  df.sort_values('Year', inplace=True)
  years = df['Year'].tolist()
  gdp = df['GDP (USD)'].tolist()
  return years, gdp  

#get data for India and Pakistan
years_india, gdp_india = get_gdp_years('India')
years_pak, gdp_pak = get_gdp_years('Pakistan')

#compare gdp using line graph
plt.plot(years_india, gdp_india, label = 'India')
plt.plot(years_pak, gdp_pak, label = 'Pakistan')

plt.xlabel('Years')
plt.ylabel('GDP')
plt.legend()

plt.show()

```

```python
#comparing gdp_per_capita of India and Pakistan

def get_gdpcap_years(country):
  df = excel_data.loc[excel_data['Country Name'] == country, ['Year', 'GDP per capita (USD)']]
  df.sort_values('Year', inplace=True)
  years = df['Year'].tolist()
  gdpcap = df['GDP per capita (USD)'].tolist()
  return years, gdpcap  

#get data for India and Pakistan
yearscap_india, gdpcap_india = get_gdpcap_years('India')
yearscap_pak, gdpcap_pak = get_gdpcap_years('Pakistan')

#compare gdp using line graph
plt.plot(yearscap_india, gdpcap_india, label = 'India')
plt.plot(yearscap_pak, gdpcap_pak, label = 'Pakistan')

plt.xlabel('Years')
plt.ylabel('GDP per Capita')
plt.legend()

plt.show()
```

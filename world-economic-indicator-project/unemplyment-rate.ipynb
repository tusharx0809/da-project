```python
#This code takes input as the country name and draws the graph of country's unemployment graph
import pandas as pd
import matplotlib.pyplot as plt

#importing the excel file
excel_data = pd.read_excel('D:\\Python\\da-project\\world-economic-indicator-project\\dataset\\WorldBank.xlsx')

def get_unemployment_rate(country):
  data = excel_data.loc[excel_data['Country Name'] == country , ['Year','Unemployment (% of total labor force) (modeled ILO estimate)']]
  data.sort_values('Year', inplace = True)
  output = {}
  for i, row in data.iterrows():
    if pd.isnull(row['Unemployment (% of total labor force) (modeled ILO estimate)']):
      continue
    else:
      output[row['Year']] = row['Unemployment (% of total labor force) (modeled ILO estimate)']
  return output

def plot_graph(data):
  years = [int(key) for key,val in data.items()]
  unemployment_rate = [val for key,val in data.items()]

  plt.plot(years, unemployment_rate, label='Unemployment rate')
  plt.show()

plot_graph(get_unemployment_rate('Canada'))
```

```python
#below code compares a country's birth and death rate from year 1960 to 2018

import pandas as pd
import matplotlib.pyplot as plt
import math
import random

#importing the excel file
excel_data = pd.read_excel('D:\\Python\\da-projects\\world-economic-indicator-project\\dataset\\WorldBank.xlsx')

def get_birth_death_rate(country):
  output = {}
  data = excel_data.loc[excel_data['Country Name'] == country, ['Year','Birth rate, crude (per 1,000 people)','Death rate, crude (per 1,000 people)']]
  data.sort_values('Year', inplace= True)
  for i, row in data.iterrows():
    output[row['Year']] = get_info(row['Year'],data)
  return output

def get_info(year, data):
  output = {}
  for i, row in data.iterrows():
    if row['Year'] == year:
      output[row['Birth rate, crude (per 1,000 people)']] = row['Death rate, crude (per 1,000 people)']
  return output

def plot_graph(data):
  years = [int(key) for key,val in data.items()] #x-axis
  b = []
  d = []
  for values in data.values():
    for i, j in values.items():
      b.append(i)
      d.append(j)
  plt.plot(years, b, label = 'Birthrate')
  plt.plot(years, d, label = 'Deathrate')
  start_x, end_x = min(years),max(years)+5
  start_y, end_y = 0, 52
  plt.xticks(range(start_x, end_x, 5))
  plt.yticks(range(start_y,end_y, 4))
  plt.legend()
  plt.show()

plot_graph(get_birth_death_rate('China'))
```
